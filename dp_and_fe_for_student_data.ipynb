```python
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import warnings
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('student.csv')
```

```python
df
```

#การจัดการกะบข้อมูลที่มัน Null

```python
df = df.dropna()
df
```

#เติมค่าด้วยค่าเฉลี่ย

```python
mean_age = df['อายุ'].mean()
mean_height = df['ส่วนสูง_cm'].mean()
mean_weight = df['น้ำหนัก_kg'].mean()
df['ส่วนสูง_cm'].fillna(mean_height, inplace=True)
df['น้ำหนัก_kg'].fillna(mean_weight, inplace=True)
df['อายุ'].fillna(mean_age, inplace=True)
df
```

#การจัดการค่าผิดปกติ อายุ กับ GPA

```python
Q1_age = df['อายุ'].quantile(0.25)
Q3_age = df['อายุ'].quantile(0.75)
IQR_age = Q3_age - Q1_age
lower_bound_age = Q1_age - 1.5 * IQR_age
upper_bound_age = Q3_age + 1.5 * IQR_age
df = df[(df['อายุ'] >= lower_bound_age) & (df['อายุ'] <= upper_bound_age)]
Q1_gpa = df['GPA'].quantile(0.25)
Q3_gpa = df['GPA'].quantile(0.75)
IQR_gpa = Q3_gpa - Q1_gpa
lower_bound_gpa = Q1_gpa - 1.5 * IQR_gpa
upper_bound_gpa = Q3_gpa + 1.5 * IQR_gpa
df = df[(df['GPA'] >= lower_bound_gpa) & (df['GPA'] <= upper_bound_gpa)]
df
```

#ทำเพศ กับ คณะเป็นมาตรฐาน

```python
def standardize_faculty(name):
    if name in ['วิทยาศาสตร์', 'วิทย์', 'Science', 'คณะวิทยาศาสตร์','Sci']:
        return 'วิทยาศาสตร์'
    elif name in ['ศิลปศาสตร์', 'ศิลป์', 'Arts','Liberal Arts']:
        return 'ศิลปศาสตร์'
    elif name in ['วิศวฯ','วิศวะ','วิศวกรรมศาสตร์']:
      return 'วิศวกรรมศาสตร์'
    else:
        return name
def standardize_sex(name):
    if name in ['ชาย','ช','M','Male']:
        return 'ชาย'
    elif name in ['ญ','หญิง','F','Female']:
        return 'หญิง'
    else:
        return name
df['เพศ'] = df['เพศ'].apply(standardize_sex)
df['คณะ'] = df['คณะ'].apply(standardize_faculty)
df
```

#แปลงข้อมูลประเภทเป็นตัวเลข

```python
data_encoded = pd.get_dummies(df, columns=['คณะ'])
print(data_encoded)
```

#การสร้าง Feature จากวันเวลา

```python
df['วันที่ลงทะเบียน'] = pd.to_datetime(df['วันที่ลงทะเบียน'])
df['ปีลงทะเบียน'] = df['วันที่ลงทะเบียน'].dt.year
df['เดือนลงทะเบียน'] = df['วันที่ลงทะเบียน'].dt.month
df['ช่วงเวลาในปี'] = pd.cut(df['เดือนลงทะเบียน'],
    bins=[0, 4, 8, 12],
    labels=['ต้นปี', 'กลางปี', 'ปลายปี']
)
df
```

#การรวม Features

```python
student_health = df.copy()
student_health['BMI'] = student_health['น้ำหนัก_kg'] / (student_health['ส่วนสูง_cm']/100)**2
student_health['สุขภาพ'] = pd.cut(student_health['BMI'],
                                   bins=[0, 18.5, 25, 30, 100],
                                   labels=['ผอม', 'ปกติ', 'อ้วน', 'อ้วนมาก'])
student_health[['ชื่อ', 'BMI', 'สุขภาพ']]
```
